# Resulting Model and Parameters

## 4-State ODE Model

\begin{align}
\frac{dR}{dt} = \frac{\beta_R I}{(k_R) + I} - \rho R\\
\frac{dm_P}{dt} = \alpha + \frac{\epsilon R}{k_m + R} - \delta m_P, \text{ where  }\alpha = \epsilon l\\
\frac{dP}{dt} = (\beta_P) (m_P) - MP - \gamma P\\
\frac{dP_m}{dt} = MP - \gamma P_m
\end{align}

## By equation

### In dR/dt

beta_R: translation rate of mRNA_R  
I: concentration of Rpa AHL -- 10 micro-molars  
k_R, none: activation constant of promoter of mRNA_R  
rho: degradation rate of R

### In dmp/dt

alpha: product of transcription rate and leak of mRNA_P  
epsilon: translation rate of mRNA_P  
l: leak of mRNA_P  
k_m: activation constant of promoter for mRNA_P  
delta: degradation rate of mRNA_p

### In dP/dt and dPm/dt

beta_P: translation rate of P  
M,: maturation rate of P  
gamma: degradation rate of P
